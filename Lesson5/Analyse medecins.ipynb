```python
effectifs = pd.read_csv('data/damir2013_2.csv' ,sep=';')
effectifs = effectifs.rename(columns={'effectifs': 'nb_gen'})
effectifs = effectifs.set_index('codep')
effectifs = effectifs.drop(['mode_exercice', 'zone_inscription', 'specialite', 'sexe'], axis=1)
# recalculons les colonnes pour avoir des numériques
effectifs['acte_affilies']=effectifs.nb_actes/effectifs.pop_affiliee
# remarque : densité de medecins pour 100000 habitants
effectifs['densite_medecins']=effectifs.nb_gen/effectifs.pop_affiliee*100000
effectifs.head(5)
```

```python
# nombre d'actes total France 2013
effectifs.nb_actes.sum()
```

### Cherchons une corrélation entre la densité de médecins et le nombre d'actes par affilié

```python
from scipy.stats import pearsonr
    
```

```python
data1 = effectifs['acte_affilies']
data2 = effectifs['densite_medecins']
```

```python
# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)
```

#### Corrélation faible...

### Cherchons une corrélation entre la densité de médecins et le revenu

```python
# ayant mis le codep en index dans les 2 dataframes, on peut join implicite
# certaines données outre mer sont manquantes -> dropna
eff_rev = effectifs.join(revenu).dropna()
eff_rev.shape
```

```python
eff_rev.head(5)
```

```python
data1 = eff_rev['revenuMed']
data2 = eff_rev['densite_medecins']
```

```python
# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)
```

#### Corrélation encore plus faible...

### Pour le fun regardons la corrélation entre le % d'imposition et le revenu

```python
data1 = eff_rev['revenuMed']
data2 = eff_rev['%imposé']
```

```python
# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)
```

### ouf enfin une corrélation attendue ;-)

### Vérifions que la population d'affiliés est cohérente par département entre les Dataframes population et effectifs

```python
pop_dep = population.set_index(['l_dpt']).melt(id_vars="dpt", var_name="tranche_age")
pop_dep = pop_dep.drop(['tranche_age'],axis=1).groupby('dpt').sum()
pop_dep.head(5)
```

```python
ecart = pop_dep.join(effectifs).value - pop_dep.join(effectifs).pop_affiliee
ecart.sum()
```

#### Les dataframes sont cohérents !
